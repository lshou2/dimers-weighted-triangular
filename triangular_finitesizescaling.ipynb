## Vison correlator plot

Note: The following calculations can take several minutes.

````python
n=102
v102al = []
for a in np.arange(2,4,.05):
    v102al.append(vd_triangular(n, square_bool, pathlength=n//2,weightlist=[a,1,1,1,1,1], xlog=False,\
                    verbose=False, plot=False, vert=True, colshift=0)[0][-1])
n=202
v202al = []
for a in np.arange(2,4,.05):
    v202al.append(vd_triangular(n, square_bool, pathlength=n//2,weightlist=[a,1,1,1,1,1], xlog=False,\
                    verbose=False, plot=False, vert=True, colshift=0)[0][-1])
n=302
v302al = []
for a in np.arange(2,4,.05):
    v302al.append(vd_triangular(n, square_bool, pathlength=n//2,weightlist=[a,1,1,1,1,1], xlog=False,\
                    verbose=False, plot=False, vert=True, colshift=0)[0][-1])
````

````python
# vison correlator plot along vertical path
pt=16
plt.plot(np.arange(2,4,.05),np.sqrt(np.array(v102al)),'.-',label=r'$N=101$')
plt.plot(np.arange(2,4,.05),np.sqrt(np.array(v202al)),'.-',label=r'$N=201$',alpha=.7)
plt.plot(np.arange(2,4,.05),np.sqrt(np.array(v302al)),'.-',label=r'$N=301$',alpha=.5)
plt.legend(fontsize=pt-2)
plt.xticks(fontsize=pt-2)
plt.yticks(fontsize=pt-2)
plt.xlabel(r'$\alpha$',fontsize=pt)
plt.ylabel(r'$V=\sqrt{|\langle v_0v_\ell\rangle|}$',fontsize=pt)
plt.savefig('sqrtv_plot.pdf', bbox_inches='tight')
````
